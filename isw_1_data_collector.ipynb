```python
import requests
import time
import os
import shutil
from calendar import monthrange
```

```python
OUTPUT_FOLDER = "data/isw/raw"

if not os.path.exists(OUTPUT_FOLDER):
   os.makedirs(OUTPUT_FOLDER)
else:
    shutil.rmtree(OUTPUT_FOLDER)
    os.makedirs(OUTPUT_FOLDER)
```

```python
def save_page_html(url, file_name):
    page = requests.get(url)
    with open(f"{OUTPUT_FOLDER}/{file_name}__report.html", "wb+") as f:
        f.write(page.content)
```

## February 25 - October 20, 2022
Format: https://isw.pub/RusCampaignFeb25

```python
year = 2022
months = ["Feb", "Mar", "Apr", "May", "June", "July", "August", "Sept", "Oct"]
BASE_URL = "https://isw.pub/RusCampaign"

for month_idx, month in enumerate(months):
    first_day = 1
    
    if month == "Feb":
        first_day = 25
    
    day_count = monthrange(year, month_idx + 2)
    
    for day in range(first_day, day_count[1] + 1):
        if month == "Oct" and day > 20:
            break
        
        file_name = f"{day}_{month_idx + 2}_{year}"
        url = f"{BASE_URL}{month}{day}"
        
        print(url)
        save_page_html(url, file_name)
        time.sleep(0.5)
```

## Redownload some exceptions
March 18, April 10, April 11, May 14, September 25, October 9

```python
year = 2022
months = ["january", "february", "march", "april", "may", "june", "july", "august", "september", "october", "november", "december"]
dates = ["18-3", "10-4", "11-4", "14-5", "25-9", "9-10"]
BASE_URL = "https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment"

for date in dates:
    month_idx = int(date.split("-")[1])
    month = months[month_idx - 1]
    day = int(date.split("-")[0])
    
    file_name = f"{day}_{month_idx}_{year}"
    url = f"{BASE_URL}-{month}-{day}"

    print(url)
    save_page_html(url, file_name)
    time.sleep(0.5)
```

## October 21 - December 31, 2022
Format: https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment-october-21

```python
year = 2022
months = ["october", "november", "december"]
BASE_URL = "https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment"

for month_idx, month in enumerate(months):
    day_count = monthrange(year, month_idx + 10)
    
    for day in range(first_day, day_count[1] + 1):
        if month == "october" and day < 21:
            continue
        
        file_name = f"{day}_{month_idx + 10}_{year}"
        url = f"{BASE_URL}-{month}-{day}"
        
        print(url)
        save_page_html(url, file_name)
        time.sleep(0.5)
```

## January 1 - January 25, 2023
Format: https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment-january-1-2023

```python
year = 2023
months = ["january"]
BASE_URL = "https://www.understandingwar.org/backgrounder/russian-offensive-campaign-assessment"

for month_idx, month in enumerate(months):
    day_count = monthrange(year, month_idx + 1)
    
    for day in range(1, 26):
        file_name = f"{day}_{month_idx + 1}_{year}"
        url = f"{BASE_URL}-{month}-{day}-{year}"
        
        print(url)
        save_page_html(url, file_name)
        time.sleep(0.5)
```

## Remove holidays
November 24 2022, December 25 2022, January 1 2023

```python
dates = ["24-11-2022", "25-12-2022", "1-1-2023"]

for date in dates:
    day = int(date.split("-")[0])
    month = int(date.split("-")[1])
    year = int(date.split("-")[2])
    
    file_name = f"{day}_{month}_{year}__report.html"
    file_path = f"{OUTPUT_FOLDER}/{file_name}"
    if os.path.exists(file_path):
        print(file_name)
        os.remove(file_path)
```
